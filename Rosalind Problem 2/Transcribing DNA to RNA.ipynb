```python
# open the DNA Sequence text file and extract data from it.
with open("DNA Sequence2.txt") as DNA_Seq:
    Data = DNA_Seq.readline().strip()
print(Data)
```

```python
#Transc#Transcribe a DNA string to an RNA string by replacing all occurrences of 'T' with 'U'.
def Transcription(DNA_seq):
    RNA = ""
    for nuc in DNA_seq:
        if nuc.upper() == "T":
            RNA += "U"
        else:
            RNA += nuc
    return RNA
```

```python
print(Transcription(Data))
```
