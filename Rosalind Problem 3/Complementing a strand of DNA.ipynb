```python
# open the DNA Sequence text file and extract data from it.
with open("DNA Sequence3.txt") as DNA_Strand:
    Data = DNA_Strand.readline().strip()
print(Data)
```

```python
def reverse_complement(DNA_Strand):
    complement = {'A': 'T', 'C': 'G', 'G': 'C', 'T': 'A'}
    return ''.join(complement[base] for base in DNA_Strand[::-1])
```

```python
# print the reverse complement of the DNA sequence(5'->3')
print(reverse_complement(Data))
```
